```python
import numpy as np
```

```python
xx, yy = np.meshgrid(np.linspace(-2.5, 2.5), np.linspace(-1.2, 1.2))
```

```python
def parabola_1(x, y):
    return (x - 1.2) ** 2 + y ** 2

def parabola_2(x, y):
    return (x + 1.2) ** 2 + y ** 2

zz = np.minimum(parabola_1(xx, yy), parabola_2(xx, yy))
```

```python
import scipy as sp
from scipy import ndimage

zz = sp.ndimage.gaussian_filter(zz, (3, 3), mode='reflect')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=zz, x=xx, y=yy, surfacecolor=np.ones_like(zz), contours=go.surface.Contours(
        z=go.surface.contours.Z(highlight=True, highlightcolor='white'),
        y=go.surface.contours.Y(highlight=False),
        x=go.surface.contours.X(highlight=False),
    ),showscale=False,hoverinfo='none',
                                )],
    layout = go.Layout(
        scene=go.layout.Scene(
            xaxis = go.layout.scene.XAxis(showspikes=False),
            yaxis = go.layout.scene.YAxis(showspikes=False),
            zaxis = go.layout.scene.ZAxis(showspikes=False),
        ), 
        showlegend=False,
    )
)
# fig.update_traces()

# fig.update_layout(title='Mt Bruno Elevation', autosize=False,
#                   width=500, height=500,
#                   margin=dict(l=65, r=50, b=65, t=90))
fig.show()
fig.write_html("saddle_demo.html")
```
